# Clase 01: Creando tu primer agente con N8N

Creando una cuenta en https://n8n.io/

N8N es una herramienta de pago con 14 días de prueba gratis.

RAG: Generación aumentada por recuperación (Retrieval-Augmented Generation)

>```
> 
> Trigger: Prompt
>     -> Agente (Memoria, herramientas, conexión API IA) <=> RAG o WEB
> 	Si es interno de la empresa => RAG
> 	Si es externo de la empresa => WEB
> 		-> Respuesta, Reporte, Markdown
> 
>```

**Paso 1:**

Crear un trigger en N8N. Se crea como Chat

**Paso 2:**

Seleccionar el modelo a usar: Geminis google, generado la API KEY en AI Studio.

**Paso 3:**

Agregar una herramienta de búsqueda web en https://serpapi.com. De igual manera se crea una cuenta y se obtiene la API KEY.

**Paso 4:**

Agregar un gestor de memoria: Para esta Inmersión se usará *Simple Memory* (For Beginners) con un *Context Window Length* de 20.

**Paso 5:**

Se agregó un RAG como herramienta. Se usó el *RAG starter template*. Se abrió una nueva pestaña donde se copio el flujo a utilizar ().
Además, se sustituyó el model IA de Embeddings por el de Gemini ya que usaba por default el de OpenAI (Figura 2).
